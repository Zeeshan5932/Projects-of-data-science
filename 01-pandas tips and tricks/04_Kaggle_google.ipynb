# Assignment: Change below variable into numeric?

1. size
2. install
3. price

```python
# Method 1: For size
df['Size'] = df['Size'].str.replace('M', '').str.replace('K', '').str.replace('Varies with device', 'NAN')
df['Size'] = df['Size'].replace("NAN",np.nan)
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')

# Method 2: For Installs
df['Installs'] = df['Installs'].str.replace(", "," ").str.replace('+', " ")
df['Installs'] = pd.to_numeric(df['Installs'], errors= 'coerce')

# Method 3: Using price() function
df['Price'] = df['Price'].str.replace('$'," ")
df['Price'] = pd.to_numeric(df['Price'], errors= 'coerce')
```

# Check that how many mbs are there in one kb? how to calculate?
1. converts kbs into m
2. then remove M from all numbers 
3. handles  `varies with device`

```python


# Replace 'varies with device' with 'Nill'
df['Size'].replace('Varies with device', 'Nill', inplace=True)

# Define conversion function
def convert_to_mb(value):
    if value.endswith('K'):
        return float(value[:-3]) / 1024
    elif value.endswith('M'):
        return float(value[:-2])
    else:
        return value

# Convert 'Size' column to string if it's not already


# Apply conversion function
df['Size'] = df['Size'].apply(convert_to_mb)
```
