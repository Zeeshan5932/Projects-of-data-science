# Data visulization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= sns.load_dataset('titanic')
```

```python
df.plot(kind='scatter' , x='age' , y='fare' , color= '#3da4cc')
```

```python
df.head()
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
plt.ylabel('Average fare')
plt.xlabel('Gender')
```

```python
#seaborn plots



# bar plot using seaborn on df
sns.barplot(data= df , x='sex' , y='age' , hue='class')
```

```python
sns.boxplot(data= df , x='sex' , y='age' , hue='class')
```

```python
print(sns.scatterplot(data= df , x='fare' , y='age' , hue='sex'))
# Flipping the axes
sns.scatterplot(data= df , y='fare' , x='age' , hue='sex')
```

```python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
