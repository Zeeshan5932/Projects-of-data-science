```python
# Import libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

# Missing values and inputing them

```python
df.isnull().sum()
```

```python
df.isnull().sum() / len(df) *100
```

```python
df['age'].mean()
```

```python
df['age'].median()
```

```python
df['age'].mode(0)
```

```python
# 1_drop the deck column
df.drop('deck', axis=1, inplace = True)
```

```python
# Filling missing value of age columns by the mean  of age
df['age'].fillna(df['age'].mean() , inplace=True)
```

```python
df.isnull().sum()
```

```python
# Fill embarked and embarked town
df['embark_town'].fillna(df['embark_town'].mode()[0] , inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0] , inplace=True)
df['binned'].fillna(df['binned'].mode()[0] , inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```
