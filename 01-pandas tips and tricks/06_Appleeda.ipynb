# Observation Set 4:

Price, Size_Bytes are important columns so it is good to fill the null values

```python
#Missing values dealing
#filling mean value of Price column in the null places of that column

df['Price']= df.Price.fillna(value=df['Price'].mean())
df['Size_Bytes'] = df.Size_Bytes.fillna(value=df['Size_Bytes'].mean())

# View fill new data again
df.isnull().sum().sort_values(ascending=False)
```

```python
df.dropna(subset=['Developer_Url'], inplace=True)
df.dropna(subset=['Released'], inplace=True)
df.dropna(subset=['App_Name'], inplace=True)
```

# 06 Create a histogram of app prices?

```python

# Filter out any null price values 
prices = df[df['Price'].notna()]['Price']

# Create a histogram
plt.hist(prices) 

# Add labels
plt.title('App Price Histogram')
plt.xlabel('Price')
plt.ylabel('Count')

# Show the plot
plt.show()
```

# 07 Pivot the dataframe to show total Updated by year

```python
# Pivot the dataframe to show total Updated by year
updated_by_year = df.pivot_table(index='Updated', values='App_Id', aggfunc='count').head()
print("APP updated by year", updated_by_year)
```
